```python
def RemoveOutlier(df,var):
    q1 = df[var].quantile(0.25)
    q3 = df[var].quantile(0.75)
    IQR = q3 - q1 
    high = q3 + 1.5 * IQR 
    low = q1 - 1.5 * IQR 
    df1 = df[(df[var] <= high ) & (df[var] >=low )]
    return df1
    
```

```python
df = RemoveOutlier(df,'Salary')
```

```python
def BuildModel(X,Y):
    from sklearn.model_selection import train_test_split
    xtrain,xtest,ytrain,ytest = train_test_split(X,Y,test_size =0.25 , random_state = 13 )
    
    
    from sklearn.linear_model import LogisticRegression
    model = LogisticRegression(solver = "lbfgs")
    model = model.fit(xtrain,ytrain)

    ypred = model.predict(xtest)
    ypred_train = model.predict(xtrain)

    from sklearn.metrics import confusion_matrix
    cm  = confusion_matrix(ytest,ypred)
    sns.heatmap(cm,annot=True)
    plt.show()

    from sklearn.metrics import classification_report
    
    print("classification model: ", classification_report(ytest,ypred))



    
```

```python
def buildLinear(X,Y):
    from sklearn.linear_model import LinearRegression
    from sklearn.model_selection import train_test_split
    xtrain,xtest,ytrain,ytest= train_test_split(X,Y,test_size= 0.20,random_state= 0)
    model = LinearRegression()
    model = model.fit(xtrain,ytrain)

    ypred = model.predict(xtest)
    ypred_train = model.predict(xtrain)

  

    from sklearn.metrics import mean_absolute_error
    print(mean_absolute_error(ytest,ypred))
    print("score of model is :",model.score(xtest,ytest))

    plt.scatter(ytest,ypred ,color = "red" , label="test data ")
    plt.scatter(ytrain,ypred_train , color = "blue" , label = "train data ")
    plt.xlabel("autual value ")
    plt.ylabel("predicted value ")
    plt.title("autual vs predicted value")
    plt.legend()
    plt.show()
```

```python
buildLinear(X,Y)
```

```python
from nltk.stem import WordNetLemmatizer

# Initialize lemmatizer
lm = WordNetLemmatizer()

# Sample set of words
sam = {"crying", "hugging", "eating", "bsdkad", "eated", "cries", "terided", "slaing", "chutiyed", "chutiya", "better"}

# Lemmatize each word
for word in sam:
    root = lm.lemmatize(word, pos='v')  # specify 'v' for verb
    print(f"For word '{word}', the root is: '{root}'")
```
